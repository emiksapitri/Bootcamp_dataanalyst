```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_pj = pd.read_csv('penyakit_jantung.csv')
data_pj.head()
```

```python
group_age = data_pj.groupby('age')['age'].count()
group_age
```

```python
group_trestbps = data_pj.groupby('age')['trestbps'].sum()
group_trestbps.head()
```

```python
group_chol = data_pj.groupby('age')['chol'].sum()
group_chol.head()
```

```python
#2 plot dalam 1 frame

plt.figure(figsize=[5,4])
sns.lineplot(x = group_trestbps.index, y = group_trestbps.values, label = 'trestbps')
sns.lineplot(x = group_chol.index, y = group_chol.values, label = 'cholesterol')
plt.legend()
```

```python
#2 plot dalam 2 frame dengan pendekatan prosedural (pakai subplot) (baris, kolom, letak)

plt.subplot(121) #ada 1 baris, 2 kolom, letaknya di nomor urut pertama
sns.lineplot(x = group_trestbps.index, y = group_trestbps.values, color = 'red')
plt.xlabel('Age')
plt.ylabel('Trestbps')

plt.subplot(122)
sns.lineplot(x = group_chol.index, y = group_chol.values, color = 'green')
plt.xlabel('Age')
plt.ylabel('Chol')
```

```python
#2 plot dalam 2 frame dengan pendekatan prosedural (pakai subplot) ()

plt.subplot(121)
sns.lineplot(x = group_trestbps.index, y = group_trestbps.values, color = 'red')
plt.xlabel('Age')
plt.ylabel('Trestbps')

plt.subplot(122)
sns.lineplot(x = group_chol.index, y = group_chol.values, color = 'green')
plt.xlabel('Age')
plt.ylabel('Chol')

plt.tight_layout() #biar tidak mepet
plt.show()
```

```python
#2 plot dalam 2 frame dengan pendekatan prosedural (pakai subplot) 

plt.subplot(221)
sns.lineplot(x = group_trestbps.index, y = group_trestbps.values, color = 'red')
plt.xlabel('Age')
plt.ylabel('Trestbps')

plt.subplot(222)
sns.lineplot(x = group_chol.index, y = group_chol.values, color = 'green')
plt.xlabel('Age')
plt.ylabel('Chol')

plt.subplot(223)
sns.lineplot(x = group_trestbps.index, y = group_trestbps.values, color = 'red')
plt.xlabel('Age')
plt.ylabel('Trestbps')

plt.subplot(224)
sns.lineplot(x = group_chol.index, y = group_chol.values, color = 'green')
plt.xlabel('Age')
plt.ylabel('Chol')

plt.tight_layout() #biar tidak mepet
plt.show()
```

```python
#2 plot dalam 2 frame dengan pendekatan prosedural (pakai subplot) (baris, kolom, letak)

plt.subplot(121) #ada 1 baris, 2 kolom, letaknya di nomor urut pertama
sns.lineplot(x = group_trestbps.index, y = group_trestbps.values, color = 'red')
plt.xlabel('Age')
plt.ylabel('Trestbps')

plt.subplot(122)
sns.scatterplot(x = group_chol.index, y = group_chol.values, color = 'green')
plt.xlabel('Age')
plt.ylabel('Chol')

plt.tight_layout() #biar tidak mepet
plt.show()
```

```python
#pakai matplotlib 

fig = plt.figure()
axes = fig.add_axes([0, 0, 0.7, 0.5]) # left, bottom, width, height
axes.plot(group_trestbps.index, group_trestbps.values, color = 'red')
plt.show()
```

```python
#2 plot beda ukuran, tapi 1 frame
#pakai matplotlib 

fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1.3, 1.3]) # left, bottom, width, height
axes1.plot(group_trestbps.index, group_trestbps.values, color = 'red')
axes1.set_xlabel('Age')
axes1.set_ylabel('Trestbps')

axes2 = fig.add_axes([0.2, 0.8, 0.3, 0.3]) # left, bottom, width, height
axes2.plot(group_chol.index, group_chol.values, color = 'blue')
axes2.set_xlabel('Age')
axes2.set_ylabel('Chol')

plt.show()
```

```python
#2 plot dalam 2 frame dengan pendekatan OOP

fig, axes = plt.subplots(2,2) #baris, kolom

#baris 1 kolom 1
axes[0][0].plot(group_trestbps.index, group_trestbps.values, color='red')

#baris 1 kolom 2
axes[0][1].plot(group_trestbps.index, group_trestbps.values, color='blue')

#baris 2 kolom 1
axes[1][0].plot(group_trestbps.index, group_trestbps.values, color='magenta')

#baris 2 kolom 2
axes[1][1].plot(group_trestbps.index, group_trestbps.values, color='green')

#bisa ditambahin label x dan y kalau mau

fig.tight_layout()
plt.show()
```
