```python
#cara penulisan lain

data_employee[data_employee.department == 'Sales']
```

```python
#2 kategori pakai isin

data_employee[data_employee.department.isin(['Sales', 'Finance'])]
```

```python
data_employee[data_employee.performance_score >= 700]
```

```python
data_employee
```

```python
#replace

data_employee['age'] = data_employee.age.fillna(data_employee.age.mean())
data_employee
```

```python
data_employee['income'] = data_employee.income.fillna(data_employee.income.median())
data_employee
```

```python
#exclude outlier pakai std

upper = data_employee.performance_score.mean() + 3*data_employee.performance_score.std()
lower = data_employee.performance_score.mean() - 3*data_employee.performance_score.std()

data_employee_clean = data_employee[(data_employee.performance_score < upper) & (data_employee.performance_score > lower)]
data_employee_clean
```

```python
#exclude outlier pakai quantile

upperq = data_employee.performance_score.quantile(.99) #batas atasnya 99%
lowerq = data_employee.performance_score.quantile(.01) #batas bawah 1%

data_employee_cleanq = data_employee[(data_employee.performance_score < upperq) & (data_employee.performance_score > lowerq)]
data_employee_cleanq
```

```python
#scaling

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

data_employee['scale_performance'] = scaler.fit_transform(data_employee[["performance_score"]])
#di dalam fit transform itu double [[]] agar datanya jadi 2D

data_employee
```

```python
#splitting

data_employee.name.str.split('')
```

```python
data_employee_cleanq.name.str.split('').map(lambda var: var[0])
```

```python
data_employee_cleanq['first name'] = data_employee_cleanq.name.str.split(' ').map(lambda var: var[0])
data_employee_cleanq['last name'] = data_employee_cleanq.name.str.split(' ').map(lambda var: var[1])
data_employee_cleanq
```
