```python
#nomor 1

import numpy as np
import pandas as pd
from scipy import stats

data_lalin = pd.read_csv('Data-Penindakan-Pelanggaran-Lalu-Lintas.csv')
data_lalin.head()
```

```python
#nomor 2 nentuin numerical vs categorical var

#cek data type
data_lalin.info()
```

```python
#karena int belum tentu numerik, cek nilai unik tiap var agar lebih yakin

for col in data_lalin.columns:
    unique_values = data_lalin[col].unique()
    print(f"Unique values in column '{col}': {unique_values}")
    print()
```

simpulan:
- variabel kategorik: bulan, wilayah
- variabel numerik: bap_tilang, stop_operasi, bap_polisi, penderekan, ocp_roda_dua, ocp_roda_empat, angkut_motor
- variabel 'stop_operasi_polisi' diduga numerik tapi kurang yakin karena nilainya selalu 0 (tapi di pertanyaan lanjutan tetap dimasukkan ke numerik dulu)

```python
numerical_columns = list(data_lalin.select_dtypes(include=['int64', 'float64']).columns)
categorical_columns = list(data_lalin.select_dtypes(include=['object', 'category']).columns)

print('Variabel numerik: ', numerical_columns)
print('Variabel kategorik: ', categorical_columns)
```

```python
#nomor 3

for x in numerical_columns:
    mean = np.mean(data_lalin[x])
    median = np.median(data_lalin[x])
    print(f"Mean {x}: {mean}")
    print(f"Median {x}: {median}")
    print()
```

```python
#nomor 4
for x in numerical_columns:
    std = np.std(data_lalin[x])
    datastat = data_lalin[x].sort_values()
    q1 = np.percentile(datastat, 25)
    q2 = np.percentile(datastat, 50)
    q3 = np.percentile(datastat, 75)
    print(f"Standar deviasi {x}: {std}")
    print(f"Q1 {x}: {q1}")
    print(f"Q2 {x}: {q2}")
    print(f"Q3 {x}: {q3}")
    print()
```
