conditional selection

```python
data3 = {
    'nama' : ['agung', 'emik', 'william', 'fafa'],
    'umur' : [26, 27, 30, 27],
    'alamat' : ['dps', 'gyr', 'sby', 'dps']
}

df3 = pd.DataFrame(data3)
df3
```

```python
df3['umur'] > 26
```

```python
#panggil dgn kriteria tertentu

df3.loc[df3['umur'] > 26]
```

```python
df3.loc[df3['alamat'] == 'dps']
```

```python
#filter juga kolomnya

df3.loc[df3['umur'] > 26, 'alamat'] #[baris, kolom]
```

```python
#lebih dari 1 kolom

df3.loc[df3['umur'] > 26, ['umur','alamat']]
```

```python
#lebih dari 1 kondisi

df3.loc[(df3['umur'] > 26) & (df3['umur'] < 30)]
```

```python
#ngapus baris df (sementara)

df3.drop(3) #ngapus baris indeks 3
```

```python
df3
```

```python
#ngapus baris df (permanen)

df3.drop(3, inplace = True) #atau bisa juga: df3 = df3.drop(3)
df3
```

```python
#ngapus beberapa  baris

df3.drop([0,2], inplace = True)
df3
```

```python
#ngapus kolom

data4 = {
    'nama' : ['agung', 'emik', 'william', 'fafa'],
    'umur' : [26, 27, 30, 27],
    'alamat' : ['dps', 'gyr', 'sby', 'dps']
}

df4 = pd.DataFrame(data3)
df4
```

menambahkan data

```python
data5 = {
    'nama' : ['agung', 'emik', 'william', 'fafa'],
    'umur' : [26, 27, 30, 27],
    'alamat' : ['dps', 'gyr', 'sby', 'dps']
}

df5 = pd.DataFrame(data3)
df5
```

```python
#nambahin gaji

df5['gaji'] = [1000,2000,3000,4000]
df5
```

```python
df5['total gaji'] = df5['umur'] * df5['gaji']
df5
```

```python
#ngeganti nilai

df5['total gaji'] = df5['total gaji'] - 500 #misal anggap pajak
df5
```

```python
#rename nama kolom banyak sekaligus

df5.rename(
    columns = {
        'nama' : 'name',
        'umur' : 'age',
        'alamat': 'address',
        'gaji' : 'salary',
        'total gaji': 'total salary'
    }, inplace = True
)
df5
```

```python
#mengubah data

df5['name'].replace('fafa', 'fufu', inplace = True)
df5
```

```python
#mengubah data

df5['address'].replace(['dps', 'sby'], ['bangli', 'jkt'], inplace = True)
df5
```
