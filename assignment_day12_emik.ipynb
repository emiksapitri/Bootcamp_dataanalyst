Data Cleaning

```python
import pandas as pd
```

```python
data_pgk = pd.read_csv('penyakit_ginjal_kronik.csv')
data_pgk
```

```python
#nomor 1: cari values bernilai NaN
data_pgk.isnull().sum()
```

```python
#nomor 2: hapus values NaN pada data
data_pgk_clean = data_pgk.dropna(axis=0)
data_pgk_clean
```

```python
#nomor 3: ganti value 'yes' dengan 'Iya' pada kolom 'hipertensi' dan 'diabetes'

data_pgk_clean[['hipertensi', 'diabetes']]
```

```python
data_pgk_clean[['hipertensi', 'diabetes']] = data_pgk_clean[['hipertensi', 'diabetes']].replace('yes', 'Iya')
data_pgk_clean[['hipertensi', 'diabetes']]
```

```python
data_pgk_clean
```

```python
#nomor 4: banyak data sebelum dan setelah penghapusan

print(f'Data sebelum cleaning: {data_pgk.shape[0]} baris, {data_pgk.shape[1]} kolom')
print(f'Data setelah cleaning: {data_pgk_clean.shape[0]} baris, {data_pgk_clean.shape[1]} kolom')
```
