```python
import pandas as pd

data_netflix = pd.read_csv('netflix_titles.csv')
data_netflix.head()
```

```python
#nomor 1: set index, jadikan show_id jadi index

data_netflix = data_netflix.set_index('show_id')
data_netflix
```

```python
#nomor 2: cari data duplikat
data_netflix.duplicated()
```

```python
data_netflix.duplicated().sum()
```

```python
#nomor 3

data_netflix.drop_duplicates()
```

```python
#nomor 2: histplot

import seaborn as sns
import matplotlib.pyplot as plt

group1 = data_netflix1.groupby('release_year')['type'].count()
group1
```

```python
plt.figure(figsize=[15,4])
sns.barplot(x=group1.index, y = group1.values, data = data_netflix1)
plt.title('barplot type')
plt.show()
```
