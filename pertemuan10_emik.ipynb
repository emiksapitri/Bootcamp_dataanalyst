```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
pj = pd.read_csv('penyakit_jantung.csv')
pj
```

```python
thalach = pj['thalach']
thalach.head()
```

```python
#manual
mean_thalach = sum(thalach) / len(thalach)
mean_thalach
```

```python
#pakai np
mean_thalachnp = np.mean(thalach)
mean_thalachnp
```

```python
median_thalach = np.median(thalach)
median_thalach
```

```python
modus_thalach = stats.mode(thalach)
modus_thalach
```

```python
modus_thalach[0]
```

```python
varian_thalach = np.var(thalach)
varian_thalach
```

```python
mean_thalachnp = np.mean(thalach)
median_thalach = np.median(thalach)
modus_thalach = stats.mode(thalach)
varian_thalach = np.var(thalach)
std_thalach = np.std(thalach)
skewness_thalach = stats.skew(thalach)

print(mean_thalachnp)
print(median_thalach)
print(modus_thalach[0])
print(varian_thalach) #smakin besar var maka data cenderung jauh dari rata-rata
print(std_thalach) #menentukan data condong berkumpul di rentang nilai mana (std+mean sampai std-mean)
print(skewness_thalach) #menentukan kecondongan data
```

```python
#kuartil

datastat = thalach.sort_values()

q1 = np.percentile(datastat, 25)
q2 = np.percentile(datastat, 50)
q3 = np.percentile(datastat, 75)

iqr = q3 - q1

lower_bound = q1 - 1.5*iqr
upper_bound = q3 + 1.5*iqr

print('Q1 :', q1)
print('Q2 :', q2)
print('Q3 :', q3)
print('IQR :', iqr)
print('LB :', lower_bound)
print('UB :', upper_bound)
```

```python
#outlier detection

outlier = []

for i in thalach:
    if i < lower_bound or i > upper_bound:
        outlier.append(i)

print(outlier)
```

```python
#hapus outlier

df_thalach_baru = thalach[(thalach >= lower_bound) & (thalach <= upper_bound)]
df_thalach_baru
```
